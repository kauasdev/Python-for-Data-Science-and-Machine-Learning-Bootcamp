```python
figura = plt.figure(figsize=(20,5))

figura.add_subplot(131) # 1 linha, 3 colunas, axes 1
plt.plot(x, y, label='x e y')
plt.legend()
plt.title('Gráfico 1')

figura.add_subplot(133) # 1 linha, 3 colunas, axes 3
plt.scatter(x*2, y*0.5, label='x*2 e y*0,5', marker="*")
plt.legend()
plt.title('Gráfico 2')

figura.add_subplot(132) # 1 linha, 3 colunas, axes 2
plt.bar(x*0.4, y*2.2, label='x*0,4 e y*2,2')
plt.legend()
plt.title('Gráfico 3')

plt.savefig('graficos.png') # salvando a como imagem

plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/indexData.csv')
```

```python
df.head()
```

```python
cotacao_df = df[['Index', 'Date', 'Close']]
display(cotacao_df)
```

```python
# dfremove = cotacao_df.loc[(cotacao_df['Index'] != 'NYA')]
# display(dfremove)

nya_df = cotacao_df.loc[(cotacao_df['Index'] == 'NYA')]
display(nya_df)
nya_df = nya_df[13900:]
display(nya_df)
```

```python
xmin = list(nya_df['Date'])[0]
xmax = list(nya_df['Date'])[-1]
xmid = list(nya_df['Date'])[len(nya_df['Date'])//2]

print(xmin, xmax, xmid)
```

```python
plt.figure(figsize=(20,10))
plt.plot(
    nya_df['Date'],
    nya_df['Close'],
    label='NYA',
    ls='--',
    lw='2'
)
plt.xlabel('Date')
plt.ylabel('Close')
plt.xticks([xmin, xmid, xmax])
plt.legend()
plt.title('Teste')
plt.grid()
plt.savefig('kaggle-plot.png')
plt.show()
```
