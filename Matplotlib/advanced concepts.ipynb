```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 5,  20)
y = x * 2
```

#### Log scale

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))

ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_title('Normal Scale')

ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_yscale('log')
ax[1].set_title('Log Sacle(y)')
```

### Placement of ticks and custom labels

```python
fig1, ax1 = plt.subplots(figsize=(10, 4))

ax1.plot(x, x**2, x, x**3, lw=2)

ax1.set_xticks([1,2,3,4,5])
ax1.set_xticklabels([r'$\alpha$',r'$\beta$',r'$\gamma$',r'$\delta$',r'$\epsilon$'], fontsize=18)

yticks = [0,50,100,150]
ax1.set_yticks(yticks)
ax1.set_yticklabels([f"{y:.1f}" for y in yticks], fontsize=18)
```

### Scientific Notation

```python
fig, ax = plt.subplots(1,1)

ax.plot(x, x**2, x, np.exp(x))
ax.set_title('Scientific Notation')

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker

formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
```

### Axis number and axis label spacing

```python
matplotlib.rcParams['xtick.major.pad'] = 5
matplotlib.rcParams['ytick.major.pad'] = 5

fig, ax = plt.subplots(1,1)

ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0,50,100,150])

ax.set_title("Label and axis spacing")

ax.yaxis.labelpad = 30
ax.xaxis.labelpad = 5

ax.set_xlabel('x')
ax.set_ylabel('y')
```

### Twin axes

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, lw=2, color='b')
ax.set_ylabel(r'area $(m^3)$', fontsize=18, color='blue')

for label in ax.get_yticklabels():
    label.set_color('b')
    
ax2 = ax.twinx() # cria um novo axes compartilhando o xaxis
ax2.plot(x, x**3, lw=2, color='r')
ax2.set_ylabel(r'volume $(m^3)$', fontsize=18, color='r')
for label in ax2.get_yticklabels():
    label.set_color("red")
```

### Axes where x and y is zero

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0)) # set position of x spine to x=0
# ('data',0) -> zero
# ('axes', 0.5) -> center

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0)) # set position of y spine to y=0

xx = np.linspace(-0.75, 1, 100)
ax.plot(xx, xx**3)
```
