```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 12)
y = x ** 2.5
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(
    x, 
    y, 
    color='#FF8C00', 
    lw=2, # espessura da linha (linewidth)
    alpha=0.7, # transparencia
    ls='-.', # line style
    marker='o',
    markersize=20,
    markerfacecolor='yellow',
    markeredgewidth=3,
    markeredgecolor='red'
)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x*3, y*2.3)
ax.set_xlim([-2, 16])
ax.set_ylim([-50, 130])
```
