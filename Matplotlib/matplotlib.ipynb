```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
print(x)
print(y)
```

```python
# Functional Method
plt.plot(x,y, 'r--')
# 'r--' == color='r', ls='--'
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(121)
# 121 => 1 row | 2 columns | position 1
plt.plot(x, y, 'r--')


plt.subplot(1,2,2)
plt.plot(y,x, 'b-.')

plt.show()
```

```python
# OO

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y,'r--')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

# for current_ax in axes:
#     current_ax.plot(x,y)

plt.tight_layout() # evita a sobreposição
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(6,4), dpi=100) # largura e altura em polegadas
# figsize=(width, height)
# dpi = 100 (default)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1,figsize=(8,5))

fig.suptitle('Figure', fontsize=20)

axes[0].plot(x,y)
axes[0].set_title('Axes 0')
axes[0].set_ylabel('Y Label')
axes[0].set_xlabel('X Label')
axes[1].plot(y,x)
axes[1].set_title('Axes 1')
axes[1].set_ylabel('Y Label')
axes[1].set_xlabel('X Label')

axes[2].plot(x,y, label='X vs Y')
axes[2].plot(y,x, label='Y vs X')
axes[2].legend(loc=(0.020, 0.3)) # left, bottom

plt.tight_layout()
```
