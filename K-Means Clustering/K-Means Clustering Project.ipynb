```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('College_Data')
```

```python
g = sns.FacetGrid(
    data=df,
    hue='Private',
    palette='coolwarm',
    aspect=2
)

g.map(plt.hist, 'Outstate', bins=30, alpha=.4)
```
