```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
```

## Distribution Plots

```python
sns.displot(tips['total_bill'], kde=True)
```

```python
sns.displot(titanic['age'], kde=True)
```

```python
sns.displot(titanic['age'], kind='ecdf')
```

```python
sns.displot(titanic['age'], kind='hist')
```

```python
sns.displot(titanic['age'], kind='kde')
```

```python
displot = sns.displot(
    data=titanic,
    x='pclass',
    y='survived',
    hue='sex',
    kind='kde',
    rug=True
)
displot.set_axis_labels('pclass(1,2,3)', 'survived(0,1)')
# displot.set_titles('Titanic')
plt.title('Titanic')
```
