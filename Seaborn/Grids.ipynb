# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris.shape
```

```python
iris['species'].unique()
```

## PairGrid

Pairgrid is a subplot grid for plotting pairwise relationships in a datasets.

```python
sctt = sns.PairGrid(iris, hue='species')
sctt.map(plt.scatter)
```

```python
dis = sns.PairGrid(iris, palette='coolwarm', hue='species')
dis.map_diag(sns.histplot, kde=True) # diagonal
dis.map_upper(sns.kdeplot) # acima da diagonal
dis.map_lower(plt.scatter) # abaixo da diagonal
```

## pairplot

pairplot is a simpler version of PairGrid (you'll use quite often)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette='rainbow')
```

## Facet Grid

FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
g = sns.FacetGrid(
    tips,
    col='time',
    row='smoker'
)
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(
    tips,
    col='time',
    row='smoker'
)
g.map(plt.scatter, 'total_bill', 'tip')
```

## JointGrid

JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g = sns.JointGrid(
    tips,
    x='total_bill',
    y='tip'
)
```

```python
g = sns.JointGrid(
    tips,
    x='total_bill',
    y='tip'
)
g = g.plot(sns.regplot, sns.histplot)
```
