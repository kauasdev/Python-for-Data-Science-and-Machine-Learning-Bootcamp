```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(10)
```

```python
titanic['survived'].unique()
```

```python
titanic['pclass'].unique()
```

```python
g = sns.JointGrid(
    x='fare',
    y='age',
    data=titanic,
    palette='coolwarm'
)
g.plot(plt.scatter, sns.histplot)
```

```python
sns.jointplot(
    x='fare',
    y='age',
    data=titanic
)
```

```python
sns.displot(titanic['fare'], bins=30, kde=False, color='red')
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='coolwarm')
```

```python
sns.swarmplot(
    x='class',
    y='age',
    data=titanic,
    palette='Set2'
)
```

```python
sns.countplot(data=titanic, x='sex', palette='coolwarm')
```

```python
sns.heatmap(titanic.corr(), cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist, 'age')
```
