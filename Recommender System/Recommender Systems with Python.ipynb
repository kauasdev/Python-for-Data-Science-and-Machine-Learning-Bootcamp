## Recommender Similar Movies

Now let's create a matrix that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. Note there will be a lot of NaN values, because most people have not seen most of the movies.

```python
moviematrix = df.pivot_table(
    index='user_id',
    columns='title',
    values='rating'
)
moviematrix.head()
```

```python
ratings.sort_values('num of ratings', ascending=False).head(10)
```

Let's choose tow movies: starwars, a sci-fi movie and Liar Liar, a comedy

```python
ratings.head()
```

```python
starwars_users_rating = moviematrix['Star Wars (1977)']
liarliar_user_rating = moviematrix['Liar Liar (1997)']
```

```python
starwars_users_rating.head()
```

```python
similar_to_starwars = moviematrix.corrwith(starwars_users_rating)
similar_to_liarliar = moviematrix.corrwith(liarliar_user_rating)
```

```python
# Remove NaN values

corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head(10)
# alguns desses filmes não fazem sentido, alguns foram vistos apenas
# uma vez por alguem que viu Star Wars
```

```python
corr_liarliar.sort_values('Correlation', ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head(10)
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation', ascending=False).head() 
```

```python
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
```
