```python
def verify_date_column():
    try:
        returns['date']
    except KeyError:
        returns.reset_index(inplace=True)

def convert_datetime(x):
    date = datetime.datetime.strptime(x, '%Y-%m-%d')
    return date
        
def get_data_by_year(year:str) -> pd.Series:
    
    verify_date_column()
    
    data_str_in_datetime = returns['date'].apply(convert_datetime)

    start = datetime.datetime.strptime(f'{year}-1-1', '%Y-%m-%d')
    end = datetime.datetime.strptime(f'{year}-12-31', '%Y-%m-%d')
    
    results = list()
    for date in data_str_in_datetime:
        if date >= start and date <= end:
            results.append(True)
        else:
            results.append(False)
    
    cond = pd.Series(results)
    
    return cond
```
