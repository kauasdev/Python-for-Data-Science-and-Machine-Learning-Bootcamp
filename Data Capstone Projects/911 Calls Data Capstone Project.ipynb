```python
data = []
for time in df['timeStamp']:
    hour = time.hour
    day_of_week = time.day_of_week
    month = time.month
    
    data.append([hour, day_of_week, month])
```

```python
df[['Hour', 'Day of Week', 'Month']] = data
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```
